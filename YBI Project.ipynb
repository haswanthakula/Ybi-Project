```python
x_train = x_train/255   #normalization to get value in required range(b/w 0 & 1)
x_test = x_test/255
```

```python
# define neural network architecture

model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=x_train[0].shape))  #inputlayer  28*28=784
model.add(tf.keras.layers.Dense(1568,activation= 'relu')) #hidden layer
model.add(tf.keras.layers.Dense(10,activation= 'softmax'))  #output it is MultiClass
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=50)
```

```python
import pandas as pd   #plot loss
loss = pd.DataFrame(model.history.history['loss']).plot()
```

```python
import pandas as pd   #plot accuracy
loss = pd.DataFrame(model.history.history['accuracy']).plot()
```

```python
# evaluate on test data  
import numpy as np 
y_pred = model.predict(x_test)  #model.predict gives only the probability of classes
y_pred
```

```python
y_pred = np.argmax(y_pred,axis=1)
y_pred
```

```python
y_test
```

```python
np.unique(y_test,return_counts=True) #actual outcomes
```

```python
np.unique(y_pred,return_counts=True) #predicted outcomes 
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
